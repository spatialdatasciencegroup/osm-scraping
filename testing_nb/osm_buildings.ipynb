# OSM Building Analysis

```python
import os
import overpass
import geopandas as gpd
import lib.geoscraping as geo

from lib.osmtools import overpass_bounds
import lib.osmtools as ost
import lib.misc as m

from time import perf_counter as tstamp

# tuscaloosa
bounds = {
    'west': -87.57477, 
    'south': 33.16558, 
    'east': -87.48422, 
    'north': 33.23129
}

auth_start = tstamp()
op = overpass.API()

query_start = tstamp()
response = op.get(
    "way" + overpass_bounds(bounds) + ";(._;>;);",
    verbosity="geom",
)

filter_start = tstamp()
way_features = [f for f in response.features if f.geometry['type'] == "LineString"]

convert_start = tstamp()
gdf = gpd.GeoDataFrame.from_features(way_features)

drop_start = tstamp()
gdf.drop([k for k in gdf.columns if 'tiger' in k], axis=1, inplace=True)

bdf_extract_start = tstamp()
bdf = gdf[gdf.building.notnull()]

bdf_drop_start = tstamp()
bdf = ost.drop_empty_cols(bdf)

bdf_reset_start = tstamp()
bdf.reset_index(inplace=True)

ost.gdf_overview(gdf, "Original  (gdf)")
ost.gdf_overview(bdf, "Buidlings (bdf)")

for 
```
